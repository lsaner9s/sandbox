```python
def factorial(n):
    if n > 0:
        return n * factorial(n-1)
    return 1

print(factorial(4))
```

```python
from numpy.random import uniform

def binomial_rv(n,p): 
    trial_results = 0 
    for trial in range(n):
        U = uniform()
        if U < p:
            trial_results +=1
    return trial_results

print(binomial_rv(10, 0.5))
```

```python
def fib(n):
    ultimo, anteultimo = 1, 1
    i = 2
    while i < n:
        actual = ultimo + anteultimo
        anteultimo = ultimo
        ultimo = actual
        i += 1

    return ultimo
    
print(fib(6))
```
