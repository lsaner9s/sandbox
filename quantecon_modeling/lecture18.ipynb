```python
from random import uniform
```

```python
class ECDF:

    def __init__(self,observations):

        self.observations = observations
    
    def __call__(self, x):

        count = 0

        for X in self.observations:

            if X <= x:

                count += 1
                
        return count / (len(self.observations))
        
        
```

```python
samples = [uniform(0,1) for i in range(1000)]

F = ECDF(samples)

print(F(0.5))
```

```python
class Polynomial:

    def __init__(self, coefficients):

        self.coefficients = coefficients

    def __call__(self,x):

        result = 0

        for i in range(len(self.coefficients)):

            result += (x ** i) * self.coefficients[i]

        return result
    
    def differentiate(self):
        
        new_coeffs = []

        for i in range(len(self.coefficients)):

            new_coeffs.append(i * self.coefficients[i])
           
        
        self.coefficients = new_coeffs

        return new_coeffs
        

        
        
```

```python
P = Polynomial([1,2,3])
print(P(2))
print(P.differentiate())

```
