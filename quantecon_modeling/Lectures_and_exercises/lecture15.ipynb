```python
def prod_interno(x_vals, y_vals):
    vectors = zip(x_vals, y_vals)
    return sum([x * y for x, y in vectors])

print(prod_interno([1,2,3],[1,1,1]))

print(sum([1 for x in range(100) if x % 2 == 0]))

pairs = ((2, 5), (4, 2), (9, 8), (12, 10))
sum([x % 2 == 0 and y % 2 == 0 for x, y in pairs])
```

```python
def p(x, coeff):
    return sum([(x ** i) * a for i, a in enumerate(coeff)])

print(p(1, (2,4)))
```

```python
def upper_amount(string):
    return sum([i == i.upper() and i.isalpha() for i in string])

print(upper_amount('The Rain in Spain'))
```

```python
def contained(seq_a, seq_b):
    return all(i in seq_a for i in seq_b)
   
print(contained("ab", "cjdb"))
```

```python
def linapprox(f, a, b, n, x):
    interval_length = b - a
    number_of_interval = n - 1
    step = interval_length/number_of_interval

    point = a
    while point <= x:
       point += step
    
    x0, x1 = point - step , point
    return f(x0) + (x - x0) * (f(x1) -f (x0)) / (x1 - x0)
```
