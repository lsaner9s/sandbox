```python
import numpy as np
import matplotlib.pyplot as plt
def ts_graph(α):
    T = 200
    x = np.empty(T+1)
    x[0] = 0

    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()

    plt.plot(x)
    plt.title(f"Alpha value: {α}")
    plt.show()

ts_graph(0.9)
```

```python
aplha_values = [0,0.8,0.98]
T = 200
x = np.empty(T+1)

for α in aplha_values:
    x[0] = 0
    for t in range(T):
        x[t+1] = α * x[t] + np.random.randn()
    plt.plot(x, label=f"alpha = {α}")

plt.legend()
plt.show()
```

```python
T = 200
x = np.empty(T+1)
x[0] = 0
α = 0.9

for t in range(T):
    x[t+1] = α * np.abs(x[t]) + np.random.randn()

plt.plot(x)
plt.title(f"Alpha value: {α}")
plt.show()
```

```python
T = 200
x = np.empty(T+1)
x[0] = 0
α = 0.9
def absol (x):
    return -x if x < 0 else x
    
for t in range(T):
    x[t+1] = α * absol(x[t]) + np.random.randn()

plt.plot(x)
plt.title(f"Alpha value: {α}")
plt.show()
```

```python
n = 1000000

count = 0
for i in range(n):

    #Random points between 0 and 1, so put together random (x,y) in the
    # 1x1 square centered on (0.5,0.5)
    x, y = np.random.uniform(), np.random.uniform()

    #The distance between the random (x,y) points and the center of the circle (0.5,0.5)
    d = np.sqrt((x - 0.5)**2 + (y - 0.5)**2)

    #If said distance is greater than 0.5, then the point is outside the circle as said circle
    #has a radius of 0.5.
    if d < 0.5:
        count += 1

#Since it is a 1x1 square, the area of the circle is exactly the amount of random points that fell inside it
#(count)/n = probability of landing in circle = area of circle  = count/n
area_aprox = count / n

#A = pi * r^@2 => A = pi * (0.5)^2 => A = pi * 0.2 => A = pi/4 => pi = A * 4
print(area_aprox * 4)
```
