```python
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
import numpy as np
import quantecon as qe
```

```python
M= 10000000
seed = 1234
key = jax.random.key(seed)
key, subkey = jax.random.split(key)
xi = jax.random.normal(key, (M, 20))
eta = jax.random.normal(subkey, (M, 20))

@jax.jit
def single_path_call(xi_single, eta_single):
    mu, rho, nu, = 0.0001, 0.1, 0.001
    s, h = 10, 0
    K = 100
    for t in range(20):
        vol = jnp.exp(h)
        s = s * jnp.exp(mu + vol * xi_single[t])
        h = rho * h + nu * eta_single[t]
    return jnp.maximum(s - K, 0)

m_path_call = jax.vmap(single_path_call)

@jax.jit
def mc_sv_pricer(xi_array, eta_array):
    all_payoffs = m_path_call(xi_array, eta_array)
    return jnp.mean(all_payoffs)

    

p = (0.99 ** 20) * mc_sv_pricer(xi, eta)
print(p)
```
