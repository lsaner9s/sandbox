```python
import numpy as np
from random import uniform
```

```python
def p(x, coeffs):

    powers = np.ones_like(coeffs)
    powers[1:] = x
    X_powers = np.cumprod(powers)
    X_powers = X_powers * coeffs

    return X_powers.sum()


print(p(2,[2,3,4]))
```

```python
class DiscreteRV:
    def __init__(self, q):
        self.q = q
        

        
    def draw(self,k = 1):
        draws = np.random.uniform(size=k)
        I = np.cumsum(self.q)
        index = np.searchsorted(I,draws)
        return index

d = DiscreteRV([0.25,0.75])
print(d.draw(4))
```
